# Definir función Sigmoide

$S(x)=\frac{1}{1+e^{-x}}$

```python
import math
def Sigmoide(X):
    sig=1/(1+math.e**(-X))
    return sig
```

# Definir función de Regresión logistica

```python
def Regresion_Logistica(X,y,theta,alpha,epochs):
    y=np.reshape(y,(len(y),1))
    N=len(X)
    Loss_List=[]
    for epoch in range(epochs):
        sigXtheta=Sigmoide(np.dot(X,theta))
        gradiente=(1/N)*np.dot(np.transpose(X),sigXtheta-y)
        theta=theta-alpha*gradiente
        hyp=Sigmoide(np.dot(X,theta))
        loss= -np.sum(np.dot(np.transpose(y),np.log(hyp)+np.dot(np.transpose(1-y),np.log(1-hyp))))/len(hyp)
        Loss_List.append(loss)
    plt.plot(np.arange(1, epochs), Loss_List)
    plt.title('Cost function')
    plt.xlabel('Epochs')
    plt.ylabel('Cost')
    plt.show()
    return theta
        
```
